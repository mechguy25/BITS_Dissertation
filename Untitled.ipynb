```python
import pandas as pd
df = pd.read_csv('after_cluster.csv')
```

```python
df.columns
```

```python
df_part = df[['RPTG_SLS_MGMT_LVL_1', 'DELL_INDUSTRY_TAXONOMY_L1',
       'ORG_USER_DEFINED_VALUE_113', 'emp_size', 'buy_power', 'decision_maker']]
```

```python
df_part.shape
```

```python
df_part.dtypes
```

```python
df_part_enc = pd.get_dummies(df_part, columns=['RPTG_SLS_MGMT_LVL_1','DELL_INDUSTRY_TAXONOMY_L1','ORG_USER_DEFINED_VALUE_113'])
```

```python
from sklearn.cluster import KMeans
```

```python
df_part_enc = df_part_enc.fillna('0')
```

```python
km=KMeans(n_clusters=4)
y_predicted = km.fit_predict(df_part_enc)
y_predicted
```

```python
df['time_taking'] = pd.DataFrame(y_predicted)
```

```python
df.groupby(['k-means']).count()
```

```python
df.to_csv('after_cluster.csv')
```
